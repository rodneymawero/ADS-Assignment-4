```python
# Select demographic columns and churn
df = df[['customerID', 'gender', "SeniorCitizen", 'Partner', 'Dependents', "Churn"]]
```

```python
df
```

```python
# define churn variables
churn_no = df[df['Churn'] == 'No']
churn_yes = df[df['Churn'] == 'Yes']
```

```python
churn_no.head()
```

```python
churn_yes.head()
```

```python
# Categorical plot (catplot) 1 for visualizing the count of customers not churned by gender
demo_no = churn_no.groupby(['gender'])['customerID'].count().reset_index()

g = sns.catplot(data = demo_no, x = 'gender', y = 'customerID',  hue = 'gender', kind = 'bar')
# extract the matplotlib axes_subplot objects from the FacetGrid
ax = g.facet_axis(0, 0)  # or ax = g.axes.flat[0]

# iterate through the axes containers
for c in ax.containers:
    labels = [f'{(v.get_height() / 1000):.1f}K' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge')
```

```python
# Males customers are slightly more than females who still did business with the telco.
```

```python
# Catplot for visualizing the count of customers not churned grouped by gender and whether they are Senior Citizens
demo_no = churn_no.groupby(['SeniorCitizen', 'gender'])['customerID'].count().reset_index()

g = sns.catplot(data = demo_no, x = 'SeniorCitizen', y = 'customerID',  hue = 'gender', kind = 'bar')
# extract the matplotlib axes_subplot objects from the FacetGrid
ax = g.facet_axis(0, 0)  # or ax = g.axes.flat[0]

# iterate through the axes containers
for c in ax.containers:
    labels = [f'{(v.get_height() / 1000):.1f}K' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge')
```

```python
# Most customers not churned are not Senior Citizens 4.5K (0). Senior Citizens are very few 0.6K
```

```python
# Grouping by partners and gender
demo_no = churn_no.groupby(['Partner', 'gender'])['customerID'].count().reset_index()

g = sns.catplot(data = demo_no, x = 'Partner', y = 'customerID',  hue = 'gender', kind = 'bar')
# extract the matplotlib axes_subplot objects from the FacetGrid
ax = g.facet_axis(0, 0)  # or ax = g.axes.flat[0]

# iterate through the axes containers
for c in ax.containers:
    labels = [f'{(v.get_height() / 1000):.1f}K' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge')
```

```python
# The numbers of those who did not churn and had partners is higher (2.8K) than those without partners (2.5K). 
```

```python
# Grouping by dependents and gender
demo_no = churn_no.groupby(['Dependents', 'gender'])['customerID'].count().reset_index()

g = sns.catplot(data = demo_no, x = 'Dependents', y = 'customerID',  hue = 'gender', kind = 'bar')
# extract the matplotlib axes_subplot objects from the FacetGrid
ax = g.facet_axis(0, 0)  # or ax = g.axes.flat[0]

# iterate through the axes containers
for c in ax.containers:
    labels = [f'{(v.get_height() / 1000):.1f}K' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge')
```

1 a. Which 2 demographics have proven to be the most popular with the telco brand?
    1. Those who are not Senior Citizens (4.5 K)
    2. Those who do not have dependents  (3.4 K)

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df = df[['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 
            'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'MonthlyCharges','Contract', 'Churn']]
```

```python
df
```

```python
services = df[['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 
            'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']]


for service in services:
    plt.figure(figsize = (12, 8))
    sns.barplot(x = service, y = 'MonthlyCharges', data = df, hue = 'Churn', )
    plt.title(service)
    plt.show()
```

b. Should the company strictly go paperless for their monthly billings as a technique
of keeping their customers and why?
  # No. The company should not completely go paperless because its loyal customers prefer Paperless billing. Also, more customers who churn  use paperless billing.  The decision to strictly go paperless for monthly billings should be based on a comprehensive analysis of customer preferences, cost-effectiveness, environmental impact, convenience, and customer retention rates. The pros and cons should be considered before making a decision.
